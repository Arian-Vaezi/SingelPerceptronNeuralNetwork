```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os
sys.path.insert(0, os.path.abspath(".."))
from src.perceptron import Perceptron

def main():
    data = np.loadtxt('../data/simple_lr.csv', delimiter=',',skiprows=1)
    X=data[:,:2]
    y=data[:,2]

    p= Perceptron(n_inputs=2,learning_rate=0.15)
    loss_history=p.train(X,y,epochs=200)
    print(f"Learned weight:{p.weights}   bias:{p.bias}")
    predict=p.predict(X)
    print(predict)

    plt.plot(loss_history)
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.title('Training Curve')
    plt.show()

    
if __name__ == '__main__':
    main()
```

```python

import sys, os
sys.path.insert(0, os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt
from src.perceptron import Perceptron


np.random.seed(42)
X = np.linspace(-1.0, 1.0, 1000).reshape(-1, 1)   # shape (1000,1)
noise = np.random.randn(1000) * 3
y = 2 * X.flatten() + 1 + noise

p = Perceptron(n_inputs=1, learning_rate=0.02)    # keep LR small!
loss_history = p.train(X, y, epochs=500)

plt.figure(figsize=(8, 5))


plt.scatter(X, y, s=10, alpha=0.5, label="Data points")

x_line = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_line = p.predict(x_line)
print(f"Learned weight:{p.weights}   bias:{p.bias}")


plt.plot(
    x_line,
    y_line,
    color="red",
    linewidth=2,
    label=f"Fit: y = {p.weights[0]:.2f}x + {p.bias:.2f}"
)


plt.xlabel("x")
plt.ylabel("y")
plt.title("Perceptron Fit on Noisy Linear Data")
plt.legend()
plt.show()
```
